## 15.2 Join DataFrames
* Inner Join
> It returns a dataframe with only those rows that have common characteristics.
* Outer Join
> It returns all those records which either have a match in the left or right dataframe.
* Left Join
> It returns a dataframe containing all the rows of the left dataframe.
* Right Join
> It returns a dataframe containing all the rows of the Right dataframe.


<img src="join.png" alt="Drawing" style="width: 350px;"/> 


```python
# creating two dataframe left and right
left = pd.DataFrame({'Roll_No': [1, 2, 3, 4,5],
                         'Name': ['John', 'Herry', 'Mickel', 'tom','cherry'],
                         'Class': ['B0', 'B1', 'B2', 'B3','B4']})
right = pd.DataFrame({'Roll_No': [1, 2, 4, 6,8,9],
                          'Marks': [60, 68, 30, 82, 85,56]})
```

```python
left
```

```python
right
```

### Inner Join

```python
df_inner = pd.merge(left, right, on='Roll_No', how='inner')
```

```python
df_inner
```

### Outer Join

```python
df_outer = pd.merge(left, right, on='Roll_No', how='outer')
```

```python
df_outer
```

### Left Join

```python
df_left = pd.merge(left, right, on='Roll_No', how='left')
```

```python
df_left
```

### Right Join

```python
df_right = pd.merge(left, right, on='Roll_No', how='right')
```

```python
df_right
```

## 16.1 Linear Equation Graph

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-2,2,100)
y = 2*x+1
plt.plot(x, y, '-r', label='y=2x+1')
plt.title('Graph of y=2x+1')
plt.xlabel('x', color='#1C2833')
plt.ylabel('y', color='#1C2833')
plt.legend(loc='upper left')
plt.show()
```

## 16.2 Cubic Equation Graph

```python
import matplotlib.pyplot as plt
import numpy as np
# 100 linearly spaced numbers
x = np.linspace(-3,3,100)
# the function, which is y = x^3 here
y = x**3
# setting the axes at the centre
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
# plot the function
plt.plot(x, y, 'g', label='y=x**3')
plt.title('Graph of y=x**3')
plt.xlabel('x', color='#1C2833')
plt.ylabel('y', color='#1C2833')
plt.legend(loc='upper left')
# show the plot
plt.show()
```

## 16.3 BAR Graph

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
n = 5
milan= (18, 18, 18, 18, 22)

fig, ax = plt.subplots()
index = np.arange(n)
bar_width = 0.35
opacity = 0.9
ax.bar(index, milan, bar_width, alpha=opacity, color='yellow',
                label='CSK')
ax.set_xlabel('Seasons')
ax.set_ylabel('IPL Points ')
ax.set_title('CSK')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(('2019','2018','2015','2014','2013'))
ax.legend()
plt.show()
```

## 16.4 Pie Graph

```python
import matplotlib.pyplot as plt

labels = 'China', 'India', 'United States', 'Indonesia','Pakistan'
sizes = [1394, 1326, 332, 267,233]
explode = (0, 0, 0, 0,0)
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%.1f%%', shadow=True, startangle=90)
ax.set_aspect('equal')
plt.show()
```
